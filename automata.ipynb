```python
import pygame
import numpy as np 

#iniciamos la ventana, definimos altura,anchura,color 
pygame.init()
width = 1000
height = 1000
screen = pygame.display.set_mode((height,width))
color = 25,25,25
screen.fill(color)

#se establece el tamaño de la celda 
cell_x = 60
cell_y = 60

dimW = width / cell_x
dimH = width / cell_y 

#aquí definiriamos el estado de las celdas, vivas = 1 y muertas = 0 
gameState = np.zeros((cell_x,cell_y))

#automata

gameState[5,3] = 1
gameState[5,4] = 1
gameState[5,5] = 1

gameState[21,21] = 1
gameState[22,22] = 1
gameState[22,23] = 1
gameState[21,23] = 1
gameState[20,23] = 1


#bucle de ejecución
while True:

    newGameState = np.copy(gameState)


    for y in range(0,cell_x):
        for x in range(0,cell_y):

            #Celculamos el número de vecinos cercanos.
            n_neigh = gameState[(x-1) % cell_x, (y-1) % cell_y] + \
                      gameState[(x)   % cell_x, (y-1) % cell_y] + \
                      gameState[(x+1) % cell_x, (y-1) % cell_y] + \
                      gameState[(x-1) % cell_x, (y)   % cell_y] + \
                      gameState[(x+1) % cell_x, (y)   % cell_y] + \
                      gameState[(x-1) % cell_x, (y+1) % cell_y] + \
                      gameState[(x)   % cell_x, (y+1) % cell_y] + \
                      gameState[(x+1) % cell_x, (y+1) % cell_y]

            #definimos la regla 1: una celula muerta con exactamente 3 vecinas vivas, "revive".
            if gameState[x,y] == 0 and n_neigh == 3:
                newGameState[x,y] == 1

            #definimos la regla 2: una celula viva con menos de 2 o mas de 3 vecinas vivas, "muere".

            elif gameState[x,y] == 1 and (n_neigh < 2 or n_neigh > 3):
                newGameState[x,y] == 0



            #creamos el poligono de cada celda a dibujar.
            poly = [((x)   * dimW, y  * dimH),
                    ((x+1) * dimW, y  * dimH),
                    ((x+1) * dimW, (y+1)  * dimH),
                    ((x)   * dimW, (y+1)  * dimH)]

            #Dibujamos la celda para cada par de x e y. 
            if newGameState[x,y] == 0:
	pygame.draw.polygon(screen,(128,128,128),poly,1)
			else:
				pygame.draw.polygon(screen,(255,255,255),poly,0)

	#actualizamos el estado 
	gameState = np.copy(newGameState)

	#actualizamos la pantalla.
	pygame.display.flip()
```
