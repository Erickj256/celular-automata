```python
import pygame
import numpy as np 
import time 



width, height = 1000, 1000

screen = pygame.display.set_mode((height,width))

color = 25,25,25
screen.fill(color)

#se establece el tamaño de la celda 
cell_x = 60
cell_y = 60

dimW = width / cell_x
dimH = height / cell_y 


class Automata:

    def __init__(self):
        """ Define los atributos del autómata y asigna valores iniciales. """
        #aquí definiriamos el estado de las celdas
        self.gameState = np.zeros((cell_x,cell_y))
        
    def datos_de_inicio(self):
        """ Método auxiliar donde se pueden asignar valores inciales para
        la simulación. """
        # crea una referncia local al atributo self.gameState
        gameState = self.gameState

        #rellenamos algunos cuadros de la cuadricula 

        gameState[5,3] = 1
        gameState[5,4] = 1
        gameState[5,5] = 1

        gameState[21,21] = 1
        gameState[22,22] = 1
        gameState[22,23] = 1
        gameState[21,23] = 1
        gameState[20,23] = 1
        
    def vecinos_cercanos(self, x, y):
        """Calcula el número de vecinos cercanos."""
        gameState = self.gameState

        return gameState[(x-1) % cell_x, (y-1) % cell_y] + \
            gameState[(x)   % cell_x, (y-1) % cell_y] + \
            gameState[(x+1) % cell_x, (y-1) % cell_y] + \
            gameState[(x-1) % cell_x, (y)   % cell_y] + \
            gameState[(x+1) % cell_x, (y)   % cell_y] + \
            gameState[(x-1) % cell_x, (y+1) % cell_y] + \
            gameState[(x)   % cell_x, (y+1) % cell_y] + \
            gameState[(x+1) % cell_x, (y+1) % cell_y]

    def actualiza_celda(self, x, y, newGameState, n_neigh):
        """ Reglas para actualizar cada celda. """
        if self.gameState[x,y] == 0 and (n_neigh == 3):
            newGameState[x,y] = 1
        elif self.gameState[x,y] == 1 and (n_neigh < 2 or n_neigh > 3):
            newGameState[x,y] = 0

    def dibuja_poli(self, x, y, newGameState):
        """ Dibuja la celda x, y """
        #creamos el poligono de cada celda a dibujar.
        poly = [((x)   * dimW, y  * dimH),
                ((x+1) * dimW, y  * dimH),
                ((x+1) * dimW, (y+1)  * dimH),
                ((x)   * dimW, (y+1)  * dimH)]

        #Dibujamos la celda para cada par de x e y. 
        if newGameState[x,y] == 0:
            pygame.draw.polygon(screen,(128,128,128),poly,1)
        else:
            pygame.draw.polygon(screen,(25,158,218),poly,0)

    def run(self):
        #bucle de ejecución

        while True:

            newGameState = np.copy(self.gameState)

            #screen.fill(color)

            time.sleep(0.1)


            for y in range(0,cell_x):
                for x in range(0,cell_y):

                    #Celculamos el número de vecinos cercanos.
                    n_neigh = self.vecinos_cercanos(x, y)

                    # En esta seccion irian las reglas, intentemos primero un caso pequeño 
                    self.actualiza_celda(x, y, newGameState, n_neigh)

                    # Dibuja la celda
                    self.dibuja_poli(x, y, newGameState)

            #actualizamos el estado 
            self.gameState = np.copy(newGameState)

            #actualizamos la pantalla.
            pygame.display.flip()

if __name__ == '__main__':
    # Este código sólo se ejecuta al correr el script

    #iniciamos la ventana, definimos altura,anchura,color 
    pygame.init()

    #creamos un autómata
    a = Automata()

    # rellenamos algunos cuadros de la cuadricula 
    # para cambiar condiciones iniciales basta con cambiar este método
    a.datos_de_inicio()

    # ejecuta simulación
    a.run()
```
